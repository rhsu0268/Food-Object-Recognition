# Exploratory Data Analysis

```python
# replace M and B with 1 and 0 for ease of analysis
df['diagnosis'].replace(['M', 'B'], [1, 0], inplace=True)
df.head()
```

```python
# check for missing values

print(df.isnull().sum())
```

```python
data = pd.DataFrame(df, columns = ['diagnosis', 'radius_mean'])
data
```

```python
plt.figure(figsize = (15, 5))
plt.scatter(x = data['radius_mean'], y = data['diagnosis'])
plt.title('Radius Mean of Cell Nuclei and Diagnosis')
plt.xlabel('Radius Mean')
plt.ylabel('Diagnosis')
plt.show()
```

```python
# Here we see that the diagnosis tends to be on the malignant side when the mean is large. 
# Once we get past a threshold of about 18, the diagnosis is all malignant. 
```

```python
# texture mean 

texture_data = pd.DataFrame(df, columns = ['diagnosis', 'texture_mean'])
texture_data.head()
```

```python
plt.figure(figsize = (15, 5))
plt.scatter(x = texture_data['texture_mean'], y = texture_data['diagnosis'])
plt.title('Texture Mean of Cell Nuclei and Diagnosis')
plt.xlabel('Texture Mean')
plt.ylabel('Diagnosis')
plt.show()
```

```python
# perimeter mean 

perimeter_data = pd.DataFrame(df, columns = ['diagnosis', 'perimeter_mean'])
perimeter_data.head()
```

```python
plt.figure(figsize = (15, 5))
plt.scatter(x = perimeter_data['perimeter_mean'], y = perimeter_data['diagnosis'])
plt.title('Perimeter Mean of Cell Nuclei and Diagnosis')
plt.xlabel('Perimeter Mean')
plt.ylabel('Diagnosis')
plt.show()
```

```python
# Here we see a trend similar to that of the radius mean. 
# Above a certain threshold, 117 - 118, the diagnosis is malignant. 
```

## Modeling Baseline

### (Not formal - This is just a first run at basic modeling just to explore the dataset)

```python
from sklearn import cross_validation
from sklearn import neighbors
from sklearn import grid_search
from sklearn import metrics
from sklearn import linear_model
from sklearn import tree

import seaborn as sns


# start with decision tree classifier 

# set features to use 
features = ['radius_mean', 'perimeter_mean']

# set target variable name
target = 'diagnosis'

# set X and Y
X = df[features]
y = df[target]

# create seperate training and test sets with 60/40 train/test split
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.4, random_state=42)


# Instantiate model using default params
tm = tree.DecisionTreeClassifier()

# Train model on training set
tm.fit(X_train, y_train)

# Evaluate accuracy of model on test set
print("Accuracy: %0.3f" % tm.score(X_test, y_test))

# Evaluate ROC AUC score of model on test set
print('ROC AUC: %0.3f' % metrics.roc_auc_score(y_test, tm.predict_proba(X_test)[:,1]))
```

```python
# An intiial run with the decision tree classifier actually yielded a good start. 
```

```python
# get confusion matrix on test set
y_pred = tm.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

ax = plt.axes()
sns.heatmap(cm_normalized, annot=True)
ax.set_ylabel('True')
ax.set_xlabel('Pred')
plt.show()

print("Confusion Matrix:")
print(cm)
```

```python
# Random Forest - first run


from sklearn import ensemble

# Instantiate model
rf = ensemble.RandomForestClassifier(n_estimators=20, class_weight="balanced")

# Train model on training set
rf.fit(X_train, y_train)
```

```python
# Evaluate model on train set
print("Accuracy: %0.3f" % rf.score(X_train, y_train))

# Evaluate model on test set
print("Accuracy: %0.3f" % rf.score(X_test, y_test))
```

```python
# Get confusion matrix on test set
y_pred = rf.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

ax = plt.axes()
sns.heatmap(cm_normalized, annot=True)
ax.set_ylabel('True')
ax.set_xlabel('Pred')
plt.show()

print("Confusion Matrix:")
print(cm)
```

```python
# Slightly better for random forest but more tuning is required. 
```
